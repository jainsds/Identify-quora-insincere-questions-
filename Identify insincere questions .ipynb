## Quora  Insincere questions classification 

```python


import numpy as np 
import pandas as pd 

```

## Loading data 

```python
data=pd.read_csv('/kaggle/input/quora-insincere-questions-classification/train.csv')
test=pd.read_csv('/kaggle/input/quora-insincere-questions-classification/test.csv')
```

## Text preprocessing

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

```python
stopset=set(stopwords.words('english'))
```

**Bag of words**

```python
vector=TfidfVectorizer(use_idf=True , lowercase=True , strip_accents='ascii',stop_words=stopset) # Bag of words 
vector.fit(data['question_text'])
X=vector.transform(data['question_text'])
```

```python
Y=data['target']
```

## Training 

```python
from sklearn.naive_bayes import MultinomialNB 
mn=MultinomialNB()
mn.fit(X,Y)
```

```python
voc=vector.transform(test['question_text']) # transform test data
```

## Prediction

```python
predict=mn.predict(voc)
print(np.count_nonzero(predict == 1)) # These are in
```
